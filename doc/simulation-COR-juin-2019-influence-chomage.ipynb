## Statu quo (COR)

Cette stratégie de pilotage est celle du COR en 2019. 

```python
simulateur = SimulateurRetraites()

analyse = simulateur.pilotageCOR()

pl.figure(figsize=(8, 8))
pl.suptitle(titre, fontsize=16)
analyse.dessineSimulation()
analyse.dessineLegende()
```

```python
analyse.dessineVariable("P")
pl.suptitle(titre)
pl.grid()
analyse.dessineLegende()
```

```python
analyse.dessineVariable("P", scenarios_indices=range(1, 5), dessine_legende=True)
pl.suptitle(titre)
pl.ylim(bottom=30.0, top=60.0)
pl.grid()
```

```python
analyse.setLabelLongs(False)
analyse.dessineVariable("P", scenarios_indices=[1, 5], dessine_legende=True)
pl.suptitle(titre)
pl.grid()
```

```python
annee = 2050
for scenario in [1, 5]:
    print(analyse.scenarios_labels[scenario])
    print("    %d: P=%.1f%%" % (annee, 100.0 * analyse.P[scenario][annee]))
```

```python
analyse.setLabelLongs(False)
analyse.dessineVariable("P", scenarios_indices=[4, 6], dessine_legende=True)
pl.suptitle(titre)
pl.grid()
```

```python
annee = 2050
for scenario in [4, 6]:
    print(analyse.scenarios_labels[scenario])
    print("    %d: P=%.1f%%" % (annee, 100.0 * analyse.P[scenario][annee]))
```
