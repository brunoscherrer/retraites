# Simulations numériques de tous les pilotages possibles

L'objectif de cet exemple est de démontrer tous les pilotages possibles du système de retraite disponibles dans le module. 

Le pilotage de base est celui fourni par la méthode `pilotageCOR`, avec les paramètres du COR. Ce pilotage est associé à un solde négatif puis positif. 

Plusieurs stratégies de pilotage peuvent être utilisées :

1. pilotageParPensionAgeCotisations
2. pilotageParSoldePensionAge
3. pilotageParSoldePensionCotisations
4. pilotageParSoldeAgeCotisations
5. pilotageParSoldeAgeDepenses
6. pilotageParSoldePensionDepenses
7. pilotageParPensionCotisationsDepenses
8. pilotageParAgeCotisationsDepenses
9. pilotageParAgeEtNiveauDeVie (sous-entendu et par solde financier)
10. pilotageParNiveauDeVieEtCotisations (sous-entendu et par solde financier)


```python
from retraites.SimulateurRetraites import SimulateurRetraites
import pylab as pl
```

## Pilotage 1 : le statu quo du COR

Cette stratégie de pilotage est celle du COR en 2019. 

```python
simulateur = SimulateurRetraites()

analyse = simulateur.pilotageCOR()

pl.figure(figsize=(6, 8))
pl.suptitle("Projections du COR", fontsize=16)
analyse.dessineSimulation()
analyse.dessineLegende()
```

## Pilotage 1 : pensions constantes, âge à 61 ans et cotisations définis

On utilise ici le niveau de cotisation du COR, on impose un ratio (pension moyenne)/(salaire moyen) égal à celui de 2020 et un âge de départ à la retraite égal à 62 ans.

```python
Pcible = simulateur.P[1][2020]  # Pension moyenne / salaire moyen
Acible = 62.0  # Age de départ en retraite

simulateur = SimulateurRetraites()
analyse = simulateur.pilotageParPensionAgeCotisations(
    Pcible=Pcible, Acible=Acible
)

pl.figure(figsize=(6, 8))
pl.suptitle("Pensions, âge et cotisations définis", fontsize=16)
analyse.dessineSimulation()
analyse.dessineLegende()
```

```python
analyse.graphique("S")
pl.ylim(bottom=-9.0)
analyse.dessineLegende()
```

```python
analyse.graphique("Depenses")
pl.ylim(top=21.0)
analyse.dessineLegende()
```

```python
print("Pensions, âge et cotisations définis")
analyse.afficheSolutionsSimulateurCOR()
```

```python
print("Equilibre financier, pension/salaire fixe et âge imposé")
analyse.afficheSolutionsSimulateurCOR()
```
