# Simulations numériques du rapport du COR de Juin 2019

```python
from retraites import SimulateurRetraites
from retraites import SimulateurAnalyse
import pylab as pl
```

## Hypothèses : conjoncture

```python
simulateur = SimulateurRetraites('../retraites/fileProjection.json')

analyse = simulateur.pilotageCOR()

pl.figure(figsize=(14,8))
pl.suptitle(u"Projections du COR (hypothèses)",fontsize=16)
for c in range(9):
    pl.subplot(3,3,c+1)
    v,V = [ (simulateur.B,'B'), (simulateur.NR,'NR'), (simulateur.NC,'NC'), \
            (simulateur.G,'G'), (simulateur.dP,'dP'), (simulateur.TCR,'TPR'), 
            (simulateur.TCS,'TPS'), (simulateur.CNV,'CNV'), (simulateur.EV,'EV') ][c]
    analyse.graphique(v,V)
pl.tight_layout(rect=[0, 0.03, 1, 0.95])    
```

## Statu quo (COR)

Cette stratégie de pilotage est celle du COR en 2019. 

```python
simulateur = SimulateurRetraites('../retraites/fileProjection.json')

analyse = simulateur.pilotageCOR()

titre = 'Projections du COR (Juin 2019)'
pl.figure(figsize=(8,8))
pl.suptitle(titre,fontsize=16)
analyse.graphiques()
```

```python
analyse.graphique(analyse.RNV,"RNV",draw_legend=True, scenarios_indices=range(1,5))
pl.suptitle(titre);
```

```python
analyse.graphique(analyse.Depenses,"Depenses")
analyse.plot_legend()
```

```python
analyse.graphique(analyse.Depenses,"Depenses",draw_legend=True,scenarios_indices=range(1,5))
pl.plot([2000,2070],[14.0]*2, linewidth=4.0);
pl.legend(["+1,8%", "+1,5%", "+1,3%", "+1%","Règle d'or imaginée"]);
```

```python
analyse.graphique(analyse.Depenses,"Depenses",draw_legend=True,scenarios_indices=range(1,5))
s = 3
regleor_reelle = [ 100*analyse.Depenses[s][a] for a in analyse.annees ]
pl.plot(analyse.annees,regleor_reelle, linewidth=4.0);
pl.legend(["+1,8%", "+1,5%", "+1,3%", "+1%","Règle d'or étude impact"]);
```
